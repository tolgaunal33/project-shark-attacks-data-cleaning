```python
shark_df['Activity'] = shark_df['Activity'].fillna('Unknown')
```

```python
def activity_categories(Activity):

    Activity = str(Activity).lower()


    if ('swim' in Activity or 'bath' in Activity or 'wading' in Activity or 
        'standing' in Activity or 'treading water' in Activity or 'float' in Activity):
        return 'Swimming' 
    elif 'surf' in Activity:
        return 'Surfing'
    elif 'diving' in Activity or 'snorkel' in Activity or 'freediving' in Activity:
        return 'Diving'
    elif 'fish' in Activity or 'wade fishing' in Activity or 'lobstering' in Activity:
        return 'Fishing'
    elif ('body boarding' in Activity or 'boogie boarding' in Activity or 
          'foil boarding' in Activity or 'skimboard' in Activity or 'paddle' in Activity):
        return 'Board Sports'
    elif 'unknown' in Activity:
        return 'Unknown'
    else:
        return 'Other Activities'
```

```python
shark_df['Activity'] = shark_df['Activity'].apply(activity_categories)
```

```python
blue_colors = ['#003f5c', '#2f4b7c', '#376a89', '#4682b4', '#5dade2', '#4fa3a3', '#33a1a1']
```

```python
activity_count = shark_df['Activity'].value_counts()

plt.pie(activity_count,  labels=activity_count.index, colors=blue_colors, autopct='%1.1f%%', shadow=True)
plt.title('Activity Overview')
plt.show()
```

```python
shark_df['Country'] = shark_df['Country'].str.strip().str.title()
```

```python
countries_to_analyze = ['Australia', 'Mexico', 'South Africa', 'Portugal']
activity_analysis = (shark_df[shark_df['Country'].isin(countries_to_analyze)]
.groupby(['Country', 'Activity'])
.size()
.unstack(fill_value=0))

activity_analysis
```

```python
color_palette = sns.color_palette('vlag')
activity_analysis.plot(kind='bar',color=blue_colors, xlabel='Countries', ylabel='Activity Count')
plt.title('Impacting Activities per Country')
```

```python
shark_filtered = shark_df[shark_df['Country'].isin(countries_to_analyze)]
shark_filtered['Month'] = shark_filtered['Date'].dt.month
shark_filtered['Year'] = shark_filtered['Date'].dt.year
monthly_activity_counts = shark_filtered.groupby(['Year', 'Month', 'Activity']).size().reset_index(name='Count')
color_palette= sns.set_theme(style = 'whitegrid')
plt.figure(figsize=(12, 8))
sns.lineplot(data=monthly_activity_counts, x='Month', y='Count', hue='Activity', marker='o', ci=None)
plt.xlabel('Year')
plt.ylabel('Number of Shark Attacks')
plt.title('Shark Attacks by Activity Over Time')
plt.grid(True)
plt.show()
```
