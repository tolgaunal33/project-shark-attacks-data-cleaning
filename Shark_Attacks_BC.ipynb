```python

data = df[['Country', 'Injury']]

injuries_by_country = data.groupby('Country').count()
injuries_by_country.rename(columns={'Injury': 'Number of Injuries'}, inplace=True)
injuries_by_country = injuries_by_country.sort_values(by='Number of Injuries', ascending=False)

print(injuries_by_country)
```

```python
df_countries = ['MEXICO', 'AUSTRALIA', 'SOUTH AFRICA', 'PORTUGAL']
attacks = (
    df[df['Country'].isin(df_countries)]
    .groupby(['Country', 'Activity','Year'])[['Injury']]
    .count()
    .reset_index()
)

attacks
```

```python
attacks = attacks[(attacks['Injury'] > 5) & (attacks['Activity'] == 'Surfing')]
attacks = attacks.sort_values(by='Injury', ascending=False)
attacks
```

```python
import matplotlib.pyplot as plt

attacks['Year'] = attacks['Year'].astype(int)
injuries_by_year_country = attacks.groupby(['Year', 'Country'])['Injury'].sum().unstack(fill_value=0)

injuries_by_year_country.plot(kind='bar', stacked=True)
plt.title('Injuries bye year & country')
plt.xlabel('Year')
plt.ylabel('# of injuries')
plt.legend(title='Country')
plt.show()
```
